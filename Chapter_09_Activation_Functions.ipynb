# Chapter 9 — Modeling Probabilities and Nonlinearities

## Learning Objectives

This notebook covers activation functions, ReLU, sigmoid, tanh, softmax, output probabilities, and activation derivatives.

## Theoretical Explanation

Activation functions introduce nonlinearity. Without nonlinear activation functions, stacked layers would behave like one linear layer. ReLU is common in hidden layers because it is simple and efficient. Sigmoid and softmax are common when outputs need to be interpreted as probabilities.

Backpropagation also needs derivatives of activation functions so that error signals can be correctly adjusted as they pass backward through the network.

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

## 1. Common Activation Functions

```python
x = np.linspace(-5, 5, 200)
sigmoid = 1/(1+np.exp(-x))
tanh = np.tanh(x)
relu = np.maximum(0, x)
plt.plot(x, sigmoid, label='sigmoid')
plt.plot(x, tanh, label='tanh')
plt.plot(x, relu, label='relu')
plt.legend()
plt.title('Activation functions')
plt.grid(True)
plt.show()
```

### Output Interpretation

ReLU passes positive values and blocks negative values. Sigmoid maps inputs to 0–1, while tanh maps inputs to -1–1.

## 2. Softmax Probabilities

```python
def softmax(x):
    shifted = x - np.max(x)
    exps = np.exp(shifted)
    return exps / np.sum(exps)

logits = np.array([2.0, 1.0, 0.1])
probs = softmax(logits)
print('Logits:', logits)
print('Softmax probabilities:', np.round(probs, 4))
print('Sum:', probs.sum())
```

### Output Interpretation

Softmax converts raw scores into probabilities that sum to one. The highest logit receives the highest probability.
