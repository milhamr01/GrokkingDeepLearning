## Theoretical Explanation

Sequences such as sentences have variable length and order. Bag-of-words or averaged embeddings lose order information. Recurrent neural networks process one token at a time while maintaining a hidden state. The hidden state acts as a compressed memory of previous inputs.

A recurrent layer reuses the same transition weights at each time step. This makes it possible to process arbitrary-length sequences, but long-range memory can still be difficult because gradients can vanish or explode.

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

## 1. Simple Recurrent Forward Pass

```python
vocab = {'i':0, 'like':1, 'deep':2, 'learning':3}
embed = np.eye(len(vocab))
Wxh = np.random.randn(4, 5) * 0.2
Whh = np.random.randn(5, 5) * 0.2
Why = np.random.randn(5, 4) * 0.2
sentence = ['i', 'like', 'deep', 'learning']
h = np.zeros(5)
for word in sentence:
    x = embed[vocab[word]]
    h = np.tanh(x.dot(Wxh) + h.dot(Whh))
logits = h.dot(Why)
probs = np.exp(logits - logits.max())
probs = probs / probs.sum()
print('Final hidden state:', np.round(h, 3))
print('Next-word probabilities:', np.round(probs, 3))
```

### Output Interpretation

The hidden state changes as each word is processed. The final state summarizes the sequence and is used to predict an output distribution.
