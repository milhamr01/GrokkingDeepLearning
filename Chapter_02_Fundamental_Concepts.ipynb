## Theoretical Explanation

Machine learning allows systems to learn patterns instead of being explicitly programmed for every rule. Supervised learning learns a transformation from input data to known output labels. Unsupervised learning finds structure when no target label is provided. Deep learning is primarily based on neural networks and is usually parametric: the learned knowledge is stored in weights.

The important mental model is that supervised learning transforms what is known into what is desired, while unsupervised learning discovers groups or representations. Parametric models learn by adjusting a fixed set of parameters, while nonparametric models can grow their stored knowledge with the data.

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

## 1. Supervised vs. Unsupervised Toy Examples

```python
# Supervised: learn a mapping from x to y
x = np.array([1, 2, 3, 4])
y = 2 * x
weight = 0.0
alpha = 0.05
for _ in range(60):
    pred = x * weight
    delta = pred - y
    weight -= alpha * np.mean(delta * x)
print('Learned supervised weight:', round(weight, 3))
print('Predictions:', np.round(x * weight, 2))

# Unsupervised: group points by proximity to two centroids
points = np.array([[0,0], [0.2,0.1], [3,3], [3.2,2.8]])
centroids = np.array([[0,0], [3,3]])
distances = np.linalg.norm(points[:, None, :] - centroids[None, :, :], axis=2)
labels = np.argmin(distances, axis=1)
print('Unsupervised cluster labels:', labels)
```

### Output Interpretation

The supervised example learns a weight that maps input to output. The unsupervised example assigns cluster labels based on distance without using target labels.
