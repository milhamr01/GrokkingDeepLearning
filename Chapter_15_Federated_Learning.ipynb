## Theoretical Explanation

Federated learning trains a shared model without centralizing all raw data. Each client trains locally and sends model updates rather than private data. The server aggregates updates, often by averaging them. This supports learning from distributed data while reducing direct exposure of sensitive records.

Privacy is not automatic: model updates can still leak information. Secure aggregation and encryption methods reduce this risk. The chapter introduces this privacy-centered view of deep learning and connects it to decentralized AI.

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

## 1. Federated Averaging Simulation

```python
np.random.seed(RANDOM_STATE)
clients_X = [np.random.randn(40, 2) + shift for shift in [0, 1, -1]]
true_w = np.array([[1.5], [-2.0]])
clients_y = [X.dot(true_w) + 0.1*np.random.randn(len(X),1) for X in clients_X]
global_w = np.zeros((2,1))
alpha = 0.1
for round_idx in range(20):
    client_weights = []
    for X, y in zip(clients_X, clients_y):
        w = global_w.copy()
        pred = X.dot(w)
        grad = X.T.dot(pred-y) / len(X)
        w -= alpha * grad
        client_weights.append(w)
    global_w = np.mean(client_weights, axis=0)
print('Federated learned weights:', np.round(global_w.ravel(), 3))
print('True weights:', true_w.ravel())
```

### Output Interpretation

Each client updates a local model using local data. The server averages local weights to update the global model without directly accessing client datasets.
