## Theoretical Explanation

Word embeddings represent words as vectors. A useful embedding places words with similar usage patterns near each other in vector space. Meaning is not stored as a dictionary definition; it emerges from the loss function and the predictive task.

Averaging word vectors can create simple sentence representations, although this loses word order. Embeddings are foundational for NLP because they convert symbolic text into numerical vectors that neural networks can process.

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

## 1. Toy Word Embeddings and Similarity

```python
emb = {
    'king': np.array([0.9, 0.8, 0.2]),
    'queen': np.array([0.85, 0.75, 0.25]),
    'man': np.array([0.7, 0.2, 0.1]),
    'woman': np.array([0.65, 0.15, 0.15]),
    'apple': np.array([-0.5, 0.1, 0.8])
}
def cosine(a,b):
    return a.dot(b)/(np.linalg.norm(a)*np.linalg.norm(b))
for word in ['queen', 'man', 'apple']:
    print('cosine(king,', word + '):', round(cosine(emb['king'], emb[word]), 3))
analogy = emb['king'] - emb['man'] + emb['woman']
print('Analogy vector closest scores:')
for word, vec in emb.items():
    print(word, round(cosine(analogy, vec), 3))
```

### Output Interpretation

The toy vectors show that semantic similarity can be expressed as vector similarity. The analogy example demonstrates the intuition behind vector arithmetic in embeddings.

## 2. Averaged Sentence Vectors

```python
sentences = [['king','queen'], ['man','woman'], ['apple']]
for sentence in sentences:
    vec = np.mean([emb[w] for w in sentence], axis=0)
    print(sentence, '->', np.round(vec, 3))
```

### Output Interpretation

Averaging word vectors creates a fixed-size sentence representation. This is simple and useful, but it ignores word order.
