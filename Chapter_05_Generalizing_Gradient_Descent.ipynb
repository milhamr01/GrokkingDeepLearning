# Chapter 5 — Learning Multiple Weights at a Time

## Learning Objectives

This notebook generalizes gradient descent to multiple inputs, multiple outputs, and weight matrices.

## Theoretical Explanation

When a network has multiple inputs, each input has its own weight. The prediction is a dot product. The output delta is shared across the weights, but each weight update is scaled by its corresponding input. With multiple outputs, each output has its own delta. With multiple inputs and outputs, the weight update becomes an outer product between the input vector and output delta vector.

This chapter shows that the same learning logic scales naturally from one weight to many weights.

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

## 1. Multiple Inputs, One Output

```python
inputs = np.array([1.0, 0.5, -1.0])
weights = np.array([0.1, 0.2, -0.1])
target = 0.8
alpha = 0.1
for _ in range(30):
    pred = inputs.dot(weights)
    error = (pred - target)**2
    delta = pred - target
    weight_deltas = inputs * delta
    weights -= alpha * weight_deltas
print('Final prediction:', round(inputs.dot(weights), 4))
print('Learned weights:', np.round(weights, 3))
```

### Output Interpretation

Each weight update is proportional to both the output delta and the input value. Inputs with larger magnitude create larger updates.

## 2. Multiple Inputs and Outputs

```python
inputs = np.array([1.0, 0.5, -1.0])
weights = np.random.randn(3, 2) * 0.1
target = np.array([0.7, -0.2])
alpha = 0.1
for _ in range(60):
    pred = inputs.dot(weights)
    delta = pred - target
    weight_deltas = np.outer(inputs, delta)
    weights -= alpha * weight_deltas
print('Final prediction:', np.round(inputs.dot(weights), 3))
print('Target:', target)
print('Weight matrix shape:', weights.shape)
```

### Output Interpretation

The outer product creates one update for every input-output weight connection. This is the matrix form of multi-weight learning.
