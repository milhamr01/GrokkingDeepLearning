# Chapter 3 — Introduction to Neural Prediction: Forward Propagation

## Learning Objectives

This notebook covers single-input prediction, multiple-input prediction, multiple-output prediction, vector-matrix multiplication, stacked networks, and NumPy shape rules.

## Theoretical Explanation

Forward propagation is the process of sending input data through weights to produce predictions. With one input and one output, prediction is simple multiplication. With multiple inputs, the network computes a weighted sum, also known as a dot product. With multiple outputs, each output has its own set of weights. With stacked layers, the prediction from one layer becomes the input to the next layer.

This chapter is foundational because every later learning algorithm assumes that the forward pass is understood. If the forward pass is wrong, error calculation and backpropagation will also be wrong.

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

## 1. One Input, One Output

```python
def neural_network_single(input_value, weight):
    return input_value * weight

pred = neural_network_single(8.5, 0.1)
print('Single-input prediction:', pred)
```

### Output Interpretation

The network scales the input by the weight. The weight behaves like a sensitivity knob that determines how strongly the input affects the prediction.

## 2. Multiple Inputs and Multiple Outputs

```python
inputs = np.array([8.5, 0.65, 1.2])
weights_single_output = np.array([0.1, 0.2, 0.0])
pred_single = inputs.dot(weights_single_output)
print('Weighted sum prediction:', pred_single)

weights_multi_output = np.array([
    [0.1, 0.1, -0.3],
    [0.1, 0.2,  0.0],
    [0.0, 1.3,  0.1]
]).T
pred_multi = inputs.dot(weights_multi_output)
print('Multiple-output prediction:', np.round(pred_multi, 3))
```

### Output Interpretation

The dot product combines multiple input values into a single prediction. A weight matrix allows the same input vector to produce multiple outputs.

## 3. Stacked Forward Propagation

```python
ih_weights = np.array([[0.1, 0.2, -0.1], [-0.1, 0.1, 0.9], [0.1, 0.4, 0.1]]).T
ho_weights = np.array([[0.3, 1.1, -0.3], [0.1, 0.2, 0.0], [0.0, 1.3, 0.1]]).T
hidden = inputs.dot(ih_weights)
output = hidden.dot(ho_weights)
print('Hidden layer:', np.round(hidden, 3))
print('Output layer:', np.round(output, 3))
```

### Output Interpretation

The network first creates hidden features, then uses those hidden features to make final predictions. This is the basic structure of deep networks.
