## Learning Objectives

This notebook builds a simplified mental model of neural networks using layers as vectors, weights as matrices, and networks as correlation summarizers.

## Theoretical Explanation

As neural networks grow, drawing every individual weight becomes confusing. A clearer view is to represent each layer as a vector and each set of connections as a matrix. Forward propagation becomes a sequence of vector-matrix multiplications and nonlinear transformations.

This chapter emphasizes visualization and abstraction. Instead of thinking only about individual neurons, it is often better to think about how a whole layer transforms one representation into another.

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

## 1. Layer and Weight Shapes

```python
layer_0 = np.array([[1, 0, 1]])
weights_0_1 = np.random.randn(3, 5) * 0.1
weights_1_2 = np.random.randn(5, 2) * 0.1
layer_1 = np.maximum(0, layer_0.dot(weights_0_1))
layer_2 = layer_1.dot(weights_1_2)
print('layer_0 shape:', layer_0.shape)
print('weights_0_1 shape:', weights_0_1.shape)
print('layer_1 shape:', layer_1.shape)
print('weights_1_2 shape:', weights_1_2.shape)
print('layer_2 shape:', layer_2.shape)
```

### Output Interpretation

The shape of each weight matrix is determined by the size of adjacent layers. This shape discipline is essential when debugging neural networks.

## 2. Visualizing a Weight Matrix

```python
plt.imshow(weights_0_1, cmap='coolwarm')
plt.colorbar()
plt.title('Weight matrix connecting layer 0 to layer 1')
plt.xlabel('Hidden units')
plt.ylabel('Input units')
plt.show()
```

### Output Interpretation

The weight matrix visualization shows positive and negative sensitivities between layers. This is more scalable than drawing every connection manually.
