## Learning Objectives

This notebook covers convolution, weight reuse, kernels, local feature detection, and why convolutional layers reduce overfitting for image-like data.

## Theoretical Explanation

A convolutional layer reuses the same small set of weights across many spatial locations. This is useful for images because edges, corners, and local patterns can appear anywhere. Instead of learning a separate weight for every pixel-position relationship, convolution learns reusable kernels.

Weight sharing reduces the number of parameters and improves generalization. It also preserves spatial structure, which fully connected layers often ignore.

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

## 1. Simple 2D Convolution

```python
image = np.zeros((10,10))
image[:,5:] = 1
kernel = np.array([[-1, 1], [-1, 1]])
output = np.zeros((9,9))
for i in range(9):
    for j in range(9):
        patch = image[i:i+2, j:j+2]
        output[i,j] = np.sum(patch * kernel)
fig, ax = plt.subplots(1,3,figsize=(9,3))
ax[0].imshow(image,cmap='gray'); ax[0].set_title('Input image')
ax[1].imshow(kernel,cmap='gray'); ax[1].set_title('Kernel')
ax[2].imshow(output,cmap='gray'); ax[2].set_title('Convolution output')
for a in ax: a.axis('off')
plt.show()
```

### Output Interpretation

The kernel responds strongly at the vertical edge. This demonstrates feature detection through local weighted sums.
