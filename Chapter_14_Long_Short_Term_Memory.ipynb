## Theoretical Explanation

Basic recurrent networks struggle with long-range dependencies because repeated multiplication through time can shrink or explode gradients. LSTM networks address this using a memory cell and gates. Gates control what information is forgotten, what new information is written, and what information is exposed as hidden output.

The key idea is controlled memory. Instead of forcing all information through a single hidden state update, an LSTM has a pathway that can preserve information across many time steps.

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

## 1. One Simplified LSTM Cell Step

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.array([0.5, -0.2])
h_prev = np.array([0.1, 0.0])
c_prev = np.array([0.2, -0.1])
combined = np.r_[x, h_prev]
W = np.random.randn(4, 8) * 0.2
z = combined.dot(W)
f = sigmoid(z[:2])
i = sigmoid(z[2:4])
g = np.tanh(z[4:6])
o = sigmoid(z[6:8])
c = f * c_prev + i * g
h = o * np.tanh(c)
print('forget gate:', np.round(f, 3))
print('input gate:', np.round(i, 3))
print('new cell state:', np.round(c, 3))
print('new hidden state:', np.round(h, 3))
```

### Output Interpretation

The forget gate controls how much previous memory remains. The input gate controls how much new candidate information is written. The output gate controls what part of memory becomes hidden output.
